## <b> Data Scaling using MinMaxScalar from Scikit-learn </b>

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('./AmsPrediction.csv')
df.head()
```

## Data normalization

```python
scalar = MinMaxScaler()
df_values = df.values
df_scaled = scalar.fit_transform(df_values)
normalized_df = pd.DataFrame(df_scaled)
normalized_df.head()
```

```python
normalized_df.describe()
```

## Standardising Data

```python
from sklearn.preprocessing import StandardScaler
std_scalar = StandardScaler()
df_values = df.values
df_std = std_scalar.fit_transform(df_values)
standardized_df = pd.DataFrame(df_std)
standardized_df.head()
```

```python
standardized_df.describe()
```

## Convert Categorical Data to Numerical Values

```python
df = pd.read_csv('./stdcat.csv')
df.head()
```

```python
df.info()
```

#### Replacing values

```python
new_values = {'F': 0, 'M': 1}
df = df.replace({'Gender': new_values})
df.info()
```

```python
df.head()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
df['State'] = lb.fit_transform(df['State'])
df.head()
```

```python
df.info()
```

## One Hot Encoding

```python
df = pd.get_dummies(df, columns=['Category'], prefix = ['Cat'])
df.head()
```
